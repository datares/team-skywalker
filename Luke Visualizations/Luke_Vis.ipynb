## Clean Data

```python
rawData.koi_pdisposition.unique()
cleanData = rawData.copy()
cleanData['disposition'] = cleanData['koi_pdisposition'] == 'CANDIDATE'
cleanData.drop('koi_score', axis=1, inplace=True)
cleanData.drop([col for col in cleanData.columns if 'err' in col ], axis = 1, inplace = True)
```

### Create Plots

```python
# Helper function allowing you to export a graph
def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
corr = cleanData.drop(['kepid','rowid','kepoi_name','kepler_name'], axis = 1).corr()
fig, ax = plt.subplots(figsize=(10,10))  
ax = sns.heatmap(
    corr, 
    cmap=sns.diverging_palette(20, 220, n=200),
    vmin=-1, vmax=1, center=0,
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

save_fig('Correlation Matrix')
```
